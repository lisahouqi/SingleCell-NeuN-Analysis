```python
import os
from skimage.io import imread
import matplotlib.pyplot as plt
```

### Open the whole image as numpy array

```python
working_dir = 'C:/Users/Justin/Downloads/'  # Path to image data folder

img_path = os.path.join(working_dir, 'NeuN.tif')
img_path
```

```python
img = imread(img_path)
img.shape, img.dtype
```

### Plot z-slices

```python
%matplotlib notebook
```

```python
zslice0 = img[0]

plt.imshow(zslice0, clim=[0, 150], cmap='gray')
plt.show()
```

```python
zslices = img[::8]
zslices.shape
```

```python
%matplotlib inline

plt.figure(figsize=(16, 16))
for i, zimg in enumerate(zslices):
    plt.subplot(1, 8, i+1)
    plt.imshow(zimg[2500:3000, 1700:2200], clim=[0, 150], cmap='gray')
    plt.axis('off')
    plt.title('z = {}'.format(i * 8))
plt.show()
```
